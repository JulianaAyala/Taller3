```python
for i in range(n_points):
    xo[i] = random.uniform(-r,r)
    yo[i] = random.uniform(-r,r)
    zo[i] = random.uniform(-r,r)
    radio_esf[i] = sqrt((xo[i]**2)+(yo[i]**2)+(zo[i]**2))
    
    if(radio_esf[i]<=r):
        x[i] = xo[i]
        y[i] = yo[i]
        z[i] = zo[i]
```
